```python
# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply encoding to the "name" column
dataset['name'] = label_encoder.fit_transform(dataset['name'])

# Apply encoding to the "Genre" column (if needed)
dataset['genre'] = label_encoder.fit_transform(dataset['genre'])

# Display the updated DataFrame
print(dataset[['name', 'genre',]].head())
```

```python
# converting the type into dummies

dataset = pd.get_dummies(dataset,columns=['type'],dtype=int)
dataset
```

```python
x = dataset.drop('rating',axis=1)
y = dataset['rating']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
ytrain
```

```python
from sklearn.metrics import mean_squared_error, r2_score

def prediction(model):
    model.fit(xtrain, ytrain)  # Train the model
    yp = model.predict(xtest)  # Make predictions on the test set

    # Print R² and Mean Squared Error
    print(f"Train R²: {model.score(xtrain, ytrain)}")
    print(f"Test R²: {model.score(xtest, ytest)}")
    print(f"Mean Squared Error: {mean_squared_error(ytest, yp)}")
    print(f"R² Score: {r2_score(ytest, yp)}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and use a regression model
model = RandomForestRegressor(random_state=42)
prediction(model)
```

# Interview Question

## 1) Can you explain the difference between user-based and item-based collaborative filtering?
- User-based and Item-based Collaborative Filtering are two approaches used in recommendation systems to suggest items based on past interactions
1) User-Based Collaborative Filtering
How it works: It finds users similar to the target user based on their preferences or behaviors (e.g., ratings or purchases). Recommendations are made by identifying items liked by these similar users.
2) Item-Based Collaborative Filtering
How it works: It focuses on the similarity between items rather than users. It identifies items that are frequently liked or purchased together and recommends them to users based on their history.

## 2) What is collaborative filtering, and how does it work?
- Collaborative filtering is a recommendation technique used to suggest items (e.g., movies, products) to users based on their past interactions and the preferences of other users. The idea is that people with similar tastes or behaviors will like similar things
 * How It Works
It uses user-item interaction data, like ratings, purchases, or clicks.
Finds users with similar preferences and recommends items they liked.
 Identifies items similar to what a user liked and recommends them.
Based on similarity scores, it predicts items that a user is likely to enjoy or purchase.
